##### Sistem Persamaan Linier
Persamaan linear adalah dasar dalam aljabar dan digunakan dalam berbagai bidang seperti fisika, ekonomi, dan teknik. Memahaminya membantu dalam menyelesaikan banyak masalah kehidupan nyata, seperti perhitungan keuangan, kecepatan, dan optimasi.


##### 1. Satu Solusi (Unik)
Contoh sistem persamaan:

$$
\begin{aligned}
x + y + z &= 6 \quad \text{(Persamaan 1)} \\
2x + 3y + z &= 14 \quad \text{(Persamaan 2)} \\
0x + y + 2z &= 8 \quad \text{(Persamaan 3)}
\end{aligned}
$$

##### Matriks Augmented:

$$
\left[
\begin{array}{ccc|c}
1 & 1 & 1 & 6 \\
2 & 3 & 1 & 14 \\
0 & 1 & 2 & 8
\end{array}
\right]
$$

Langkah-langkah:
    1. Tukar baris jika perlu.
    2. Eliminasi elemen di bawah elemen utama (pivot).
    3. Bentuk eselon baris.
    4.Substitusi mundur.
Hasil akhirnya:

$$
x = 0,\quad y = 4,\quad z = 2
$$

Kesimpulan: Sistem memiliki satu solusi unik.

Link: https://www.geogebra.org/m/j7bz5sz9


##### 2. Tak Hingga Solusi
Contoh sistem persamaan:

$$
\begin{aligned}
x + y + z &= 5 \\
2x + 2y + 2z &= 10 \\
3x + 3y + 3z &= 15
\end{aligned}
$$

##### Matriks Augmented:

$$
\begin{array}{ccc|c}
1 & 1 & 1 & 5 \\
2 & 2 & 2 & 10 \\
3 & 3 & 3 & 15
\end{array}
$$

Kesimpulan: Sistem memiliki tak hingga banyak solusi karena ada baris nol yang menunjukkan persamaan redundan.

Link: https://www.geogebra.org/m/j7bz5sz9



##### 3. Tanpa Solusi
Contoh sistem persamaan:

$$
\begin{aligned}
x + y + z &= 3 \\
2x + 2y + 2z &= 6 \\
x + y + z &= 5
\end{aligned}
$$

##### Matriks Augmented:

$$
\begin{pmatrix}
1 & 1 & 1 & | & 3 \\
2 & 2 & 2 & | & 6 \\
1 & 1 & 1 & | & 5
\end{pmatrix}
$$

Kesimpulan: Sistem tidak memiliki solusi (inkonsisten).

Link: https://www.geogebra.org/m/ut7hraa7

##### Soal

1.definisi eigenvalue eigen vector 
Eigenvalue dan eigenvector adalah konsep penting dalam aljabar linear, khususnya dalam analisis matriks.
Eigenvector (vektor eigen) adalah vektor bukan nol yang, ketika dikalikan oleh suatu matriks, hasilnya adalah kelipatan skalar dari vektor itu sendiri.
Eigenvalue (nilai eigen) adalah skalar (bilangan) yang merupakan faktor pengali dari vektor eigen tersebut.


2.mencari eigenvalue dan eigen vector  
Langkah-langkah untuk mencari eigenvalue dan eigenvector dari suatu matriks A:

a. Mencari Eigenvalue
Untuk mencari eigenvalue, gunakan rumus:

$\det(A - \lambda I) = 0$

Langkah-langkah:

Buat matriks$A - \lambda I$
Hitung determinannya

Selesaikan persamaan karakteristik untuk mencari nilai $\lambda$(eigenvalue)
b. Mencari Eigenvector
Untuk masing-masing nila $\lambda$ substitusi ke persamaan 
$$(A - \lambda I)\mathbf{v} = 0$$
Contoh untuk$$\lambda = 3:$$

$$\begin{bmatrix}
2 & 1 \\1 & 2 \\
\end{bmatrix}\begin{bmatrix}
-1 & 1 \\1 & -1
\end{bmatrix} \begin{bmatrix}
1 \\1\end{bmatrix}$$

3. eigen vector ortonormal menjadikan eigen vector satu kesatuan

-Vektor-vektor saling tegak lurus
-Vektor-vektor memiliki panjang 1
Untuk menjadikan eigenvector ortonormal, kita:

Normalisasi vektor eigen:
Misalnya, dari eigenvector $\mathbf{v} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$panjangnya:

$$\|\mathbf{v}\| = \sqrt{1^2 + 1^2} = \sqrt{2}$$ 
vektor ortonormal:
$$\frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}$$

Jika terdapat lebih dari satu eigenvector, kita pastikan mereka ortogonal. Jika belum ortogonal, kita bisa menggunakan metode Gram-Schmidt untuk mengortogonalisasi, lalu normalisasi hasilnya.







### Dot Product (Produk Titik)
Pengertian Dot product (atau produk skalar) adalah operasi antara dua vektor yang menghasilkan skalar (bukan vektor). Operasi ini sering digunakan dalam fisika dan matematika untuk menghitung sudut antara dua vektor, kerja (work), dan proyeksi vektor.

Contoh Soal dan Penyelesaian
1. Dot Product Dua Vektor (Rumus Aljabar)
   misalkan ada dua vektor
$$\mathbf{A} = \langle 3, 4 \rangle, \quad \mathbf{B} = \langle 2, -1 \rangle$$

Hitung A⋅B
Penyelesaian:
Gunakan rumus dot product aljabar:
$$\mathbf{A} \cdot \mathbf{B} = a_1 b_1 + a_2 b_2 = (3)(2) + (4)(-1) = 6 - 4 = 2$$
Jadi, hasil dot product-nya adalah 2
