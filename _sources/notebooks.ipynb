##### Sistem Persamaan Linier
Persamaan linear adalah dasar dalam aljabar dan digunakan dalam berbagai bidang seperti fisika, ekonomi, dan teknik. Memahaminya membantu dalam menyelesaikan banyak masalah kehidupan nyata, seperti perhitungan keuangan, kecepatan, dan optimasi.


##### 1. Satu Solusi (Unik)
Contoh sistem persamaan:

$$
\begin{aligned}
x + y + z &= 6 \quad \text{(Persamaan 1)} \\
2x + 3y + z &= 14 \quad \text{(Persamaan 2)} \\
0x + y + 2z &= 8 \quad \text{(Persamaan 3)}
\end{aligned}
$$

##### Matriks Augmented:

$$
\left[
\begin{array}{ccc|c}
1 & 1 & 1 & 6 \\
2 & 3 & 1 & 14 \\
0 & 1 & 2 & 8
\end{array}
\right]
$$

Langkah-langkah:
    1. Tukar baris jika perlu.
    2. Eliminasi elemen di bawah elemen utama (pivot).
    3. Bentuk eselon baris.
    4.Substitusi mundur.
Hasil akhirnya:

$$
x = 0,\quad y = 4,\quad z = 2
$$

Kesimpulan: Sistem memiliki satu solusi unik.

Link: https://www.geogebra.org/m/j7bz5sz9


##### 2. Tak Hingga Solusi
Contoh sistem persamaan:

$$
\begin{aligned}
x + y + z &= 5 \\
2x + 2y + 2z &= 10 \\
3x + 3y + 3z &= 15
\end{aligned}
$$

##### Matriks Augmented:

$$
\begin{array}{ccc|c}
1 & 1 & 1 & 5 \\
2 & 2 & 2 & 10 \\
3 & 3 & 3 & 15
\end{array}
$$

Kesimpulan: Sistem memiliki tak hingga banyak solusi karena ada baris nol yang menunjukkan persamaan redundan.

Link: https://www.geogebra.org/m/j7bz5sz9



##### 3. Tanpa Solusi
Contoh sistem persamaan:

$$
\begin{aligned}
x + y + z &= 3 \\
2x + 2y + 2z &= 6 \\
x + y + z &= 5
\end{aligned}
$$

##### Matriks Augmented:

$$
\begin{pmatrix}
1 & 1 & 1 & | & 3 \\
2 & 2 & 2 & | & 6 \\
1 & 1 & 1 & | & 5
\end{pmatrix}
$$

Kesimpulan: Sistem tidak memiliki solusi (inkonsisten).

Link: https://www.geogebra.org/m/ut7hraa7


# SOAL DAN JAWABAN 

~ Diketahu

$[
\vec{x} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}, \quad
\vec{y} = \begin{bmatrix} -1 \\ 2 \end{bmatrix}
]$

-> Nomor 1

$[
A = \begin{bmatrix} 1 & -1 \\ 2 & 3 \end{bmatrix}
]$

$[
A\vec{x} = \begin{bmatrix} 1 & -1 \\ 2 & 3 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix}
= \begin{bmatrix} 0 \\ 5 \end{bmatrix}
]$

$[
A\vec{y} = \begin{bmatrix} 1 & -1 \\ 2 & 3 \end{bmatrix} \begin{bmatrix} -1 \\ 2 \end{bmatrix}
= \begin{bmatrix} -3 \\ 4 \end{bmatrix}
]$

-> Nomor 2

$[
A = \begin{bmatrix} 2 & 0 \\ 2 & 3 \end{bmatrix}
]$

$[
A\vec{x} = \begin{bmatrix} 2 & 0 \\ 2 & 3 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix}
= \begin{bmatrix} 2 \\ 5 \end{bmatrix}
]$

$[
A\vec{y} = \begin{bmatrix} 2 & 0 \\ 2 & 3 \end{bmatrix} \begin{bmatrix} -1 \\ 2 \end{bmatrix}
= \begin{bmatrix} -2 \\ 4 \end{bmatrix}
]$

-> Nomor 5

Dari gambar diketahui transformasi berikut:

$[
\vec{e}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix} \rightarrow \begin{bmatrix} 1 \\ 2 \end{bmatrix}, \quad
\vec{e}_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix} \rightarrow \begin{bmatrix} 1 \\ 3 \end{bmatrix}
]$

Maka matriks transformasinya adalah:

$[
A = \begin{bmatrix} 1 & 1 \\ 2 & 3 \end{bmatrix}
]$

-> Nomor 6

Dari gambar diketahui transformasi berikut:

$[
\vec{e}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix} \rightarrow \begin{bmatrix} 1 \\ 1 \end{bmatrix}, \quad
\vec{e}_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix} \rightarrow \begin{bmatrix} -1 \\ 1 \end{bmatrix}
]$

Maka matriks transformasinya adalah:

$[
A = \begin{bmatrix} 1 & -1 \\ 1 & 1 \end{bmatrix}
]$
